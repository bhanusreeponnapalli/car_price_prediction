```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re

import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome(r"C:\Users\harsh\Desktop\chromedriver.exe")
driver.maximize_window()
```

```python
# Opening the homepage 
url = "https://www.cardekho.com/"
driver.get(url)

time.sleep(2)
```

```python
#drop down 
drop_down = driver.find_element_by_xpath('//*[@id="rf01"]/header/div[2]/div/div/nav/ul/li[2]')
drop_down.click()
```

```python
# Product urls
product_url = []
for i in range(0,15):  
    print("page",i)
    url=driver.find_elements_by_xpath('//div[@class="gsc_col-xs-7  carsName"]/a')       #scraping urls
    for j in url:
        product_url.append(j.get_attribute("href")) #appending the urls in product_urls list
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)

len(product_url)
```

```python
#Make empty lists
Brand = []   
manufactured_year = []
variant= []
kilometers = []
fuel = []
location = []
price = []
owners = []
```

```python
# Brand For cardekho
for i in product_url:
    driver.get(i)
    time.sleep(2)
    
    # Extracting Brand from xpath
    try:
        brand = driver.find_element_by_xpath('//div[@class="paddingBorder clearfix"]/div/h1')      
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-')    

    # Extracting varient from xpath
    try:
        vari = driver.find_element_by_xpath('//div[@class="variant-name"]')      
        variant.append(vari.text)
    except NoSuchElementException:
        variant.append('-')   
        
     # Extracting kilometers from xpath
    try:
        km = driver.find_element_by_xpath('//li[@class="gsc_col-xs-6 gsc_col-md-4"][4]/div/div[2]')      
        kilometers.append(km.text)
    except NoSuchElementException:
        kilometers.append('-')  
    
    # fuel for carsdekho
    try:
        fl = driver.find_element_by_xpath('//div[@class="gsc_col-xs-12"]/ul/li[3]/div/div[2]')      
        fuel.append(fl.text)
    except NoSuchElementException:
        fuel.append('-')   
    
    
    # Extracting location from carsdekho
    try:
        locat = driver.find_element_by_xpath('//div[@class="ucaddressline posR"]/span/span[2]')
        location.append(locat.text + ",Delhi")
    except NoSuchElementException:
        location.append('-')    
        
    
    # Extracting price from cardekho
    try:
        pz = driver.find_element_by_xpath('//div[@class="priceSection"]')      
        price.append(pz.text)
    except NoSuchElementException:
        price.append('-') 
        
     # Extracting owner from xpath
    try:
        ow = driver.find_element_by_xpath('//div[@class="gsc_col-xs-12"]/ul/li[6]/div/div[2]')      
        owners.append(ow.text)
    except NoSuchElementException:
        owners.append('-') 

    #manufactured_year for cardekho
    manufact = []
    try:
        manu = driver.find_element_by_xpath('//div[@class="paddingBorder clearfix"]/div/h1')      
        manufact.append(brand.text)
    except NoSuchElementException:
        manufact.append('-')  
    for k in manufact:
        manufactured_year.append(k[0:5])

print(len(Brand),len(manufactured_year),len(variant),len(kilometers),len(fuel),len(location),len(price),len(owners))  
```

```python
data = pd.DataFrame({'Brand':Brand,'Variant':variant,'Manufactured_year':manufactured_year,'Kilometers':kilometers,
                    'Fuel':fuel,'location':location,'price':price,'owners':owners})
data
```
