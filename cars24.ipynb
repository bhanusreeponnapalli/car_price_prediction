```python
#Product url
product_urls = []
for j in range(0,5):  # rang 5 for 1600 records(for range(2) 640 records)
    print("page",j)
    url=driver.find_elements_by_xpath('//div[@class="col-4"]/a')       #scraping urls
    for i in url:
        product_urls.append(i.get_attribute("href")) #appending the urls in product_urls list
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        
len(product_urls)
```

```python
#Make empty lists
Brand = []   
manufactured_year = []
variant= []
kilometers = []
fuel = []
location = []
price = []
owners = []
```

```python
for i in product_urls:
    driver.get(i)
    time.sleep(2)
    
    # Extracting Brand from xpath
    try:
        brand = driver.find_element_by_xpath('//div[@class="_10kZS"]/h2')      
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-')    
    
    # Extracting varient from xpath
    try:
        vari = driver.find_element_by_xpath('//div[@class="DB7s9"]/ul/li[7]/strong')      
        variant.append(vari.text)
    except NoSuchElementException:
        variant.append('-')    

    # Extracting manufactured_date from xpath
    manufactur = []
    try:
        manuf = driver.find_element_by_xpath('//div[@class="_10kZS"]/h2')      
        manufactur.append(manuf.text)
    except NoSuchElementException:
        manufactur.append('-')  
    for k in manufactur:
        manufactured_year.append(k[0:4]) 
        
    
    # Extracting kilometers from xpath
    try:
        km = driver.find_element_by_xpath('//ul[@class="_2tp8h"]/li[1]')      
        kilometers.append(km.text)
    except NoSuchElementException:
        kilometers.append('-')  
    
    # extracting fuel from cars24
    try:
        fl = driver.find_element_by_xpath('//ul[@class="_2tp8h"]/li[3]')      
        fuel.append(fl.text)
    except NoSuchElementException:
        fuel.append('-')   
    
    
    # Extracting location from cars24
    try:
        locat = driver.find_element_by_xpath('//ul[@class="_2tp8h"]/li[7]')
        location.append(locat.text + ",Delhi")
    except NoSuchElementException:
        location.append('-')    
        
    
    # Extracting price from cars24
    try:
        pz = driver.find_element_by_xpath('//div[@class="text-right g-hPT"]/strong')      
        price.append(pz.text.split()[2])
    except NoSuchElementException:
        price.append('-') 
        
     # Extracting owner from cars24
    try:
        ow = driver.find_element_by_xpath('//ul[@class="_2tp8h"]/li[5]')      
        owners.append(ow.text)
    except NoSuchElementException:
        owners.append('-') 
        
print(len(Brand),len(manufactured_year),len(variant),len(kilometers),len(fuel),len(location),len(price),len(owners))  
```
