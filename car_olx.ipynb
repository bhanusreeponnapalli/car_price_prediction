```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re

import warnings
warnings.filterwarnings('ignore')
```

# olx.com 

```python
driver = webdriver.Chrome(r"C:\Users\harsh\Desktop\chromedriver.exe")
driver.maximize_window()
```

```python
# Opening the homepage of Amazon.in
url = "https://www.olx.in/"
driver.get(url)

time.sleep(2)
```

```python
#search bar
search = driver.find_element_by_xpath('//*[@id="container"]/header/div/div/div[2]/div/div/div[2]/div/form/fieldset/div/input')
search.send_keys("cars")
button= driver.find_element_by_xpath('//div[@class="_3b3oR"]')
button.click()
```

```python
#Product url
product_urls = []
start=0
end=15
for page in range(start,end):#for loop for scrapping 3 page
    url=driver.find_elements_by_xpath('//li[@class="EIR5N"]/a')       #scraping urls
    for i in url:
        product_urls.append(i.get_attribute("href"))                        #appending the urls in product_urls list
    nxt_button=driver.find_element_by_xpath('//div[@class="JbJAl"]/button')     #scraping the list of buttons from the page
    nxt_button.click()
    time.sleep(2)

len(product_urls)
```

```python
#Make empty lists
Brand = []   
manufactured_year = []
variant= []
kilometers = []
fuel = []
location = []
price = []
owners = []
```

```python

for i in product_urls:
    driver.get(i)
    time.sleep(0.5)
    
    # Extracting Brand from xpath
    try:
        brand = driver.find_element_by_xpath('//div[@class="_35xN1"]')      
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-')    

    # Extracting varient from xpath
    try:
        vari = driver.find_element_by_xpath('//div[@class="_3tLee"]')      
        variant.append(vari.text)
    except NoSuchElementException:
        variant.append('-') 

     # Extracting driven kilometers
    try:
        km = driver.find_element_by_xpath('//div[@class="aOxkz"]/div[2]/div')      
        kilometers.append(km.text)
    except NoSuchElementException:
        kilometers.append('-') 

     # Extracting fuel from xpath
    try:
        fl = driver.find_element_by_xpath('//div[@class="aOxkz"]/div[1]/div')      
        fuel.append(fl.text)
    except NoSuchElementException:
        fuel.append('-')    

      #Location
    try:
        locat = driver.find_element_by_xpath('//div[@class="_1ThNy"][2]/div[2]/div[2]')      
        location.append(locat.text)
    except NoSuchElementException:
        location.append('-')    
  
     #Price
    try:
        pz = driver.find_element_by_xpath('//div[@class="_3FkyT"]')      
        price.append(pz.text)
    except NoSuchElementException:
        price.append('-')    

    #number of owners
    try:
        oner = driver.find_element_by_xpath('//div[@class="_1l939"]/div[1]/div[2]/div[2]')      
        owners.append(oner.text)
    except NoSuchElementException:
        owners.append('-')  

        
    #manufactured_year
    manufactured = []
    try:
        manufactur = driver.find_element_by_xpath('//div[@class="_35xN1"]')      
        manufactured.append(manufactur.text)
    except NoSuchElementException:
        manufactured.append('-')
    for j in manufactured:
        manufactured_year.append(j[-5:-1])
        
        
    
print(len(Brand),len(manufactured_year),len(variant),len(kilometers),len(fuel),len(location),len(price),len(owners))  
    
```

```python
data = pd.DataFrame({'Brand':Brand,'Variant':variant,'Manufactured_year':manufactured_year,'Kilometers':kilometers,
                    'Fuel':fuel,'location':location,'price':price,'owners':owners})
data
```

```python
data.to_csv("car_olx.csv")
```
